# Questão 4
A imagem ```babbon_lena.png``` contém uma combinação das imagens ```baboon.png``` e ```lena.png``` utilizando a máscara ```mask.png```. Desenvolva o código que efetua essa combinação.

````python
img_1 = cv2.imread('img/baboon.png')
img_2 = cv2.imread('img/lena.png')
img_3 = cv2.imread('img/baboon_lena.png')

img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)

# Altere as próximas linhas de código
# TODO

mask = cv2.imread('img/mask.png')

notMask = cv2.bitwise_not(mask)

imgH = cv2.bitwise_and(img_1, mask)

imgH2 = cv2.bitwise_and(img_2, notMask)
plt.imshow(imgH + imgH2)
plt.show()
````

````python
# Não é preciso alterar as próximas linhas de código que devem mostrar o resultado da operação.
plt.subplot(1,3,1), plt.imshow(img_1)
plt.subplot(1,3,2), plt.imshow(img_2)
plt.subplot(1,3,3), plt.imshow(img_3)
plt.show()
````

````python
# Solução
img_1 = cv2.imread('img/radial.png')

# Altere as próximas linhas de código
# TODO

x = np.linspace(-10, 10, 101)
y = np.linspace(-10, 10, 101)

xx, yy = np.meshgrid(x, y)
yy = np.sin(xx)
zz = np.sqrt(xx**2 + yy**2)

plt.imshow(zz, cmap='gray')
````
